```python
# 데이터 읽어 오기. pandas의 read_csv 사용
data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/data/tips.csv")
data
```

## 1. total_bill 과 tip 의 분포도(scatter) 시각화하기

### tip 금액에 따라 점의 크기 변화 주기, alpha 옵션으로 투명도 지정하기
### 축 label 추가하기, 플롯 title 추가하기


```python
plt.style.use(['seaborn-notebook'])
```

```python
x = data['total_bill']
y = data['tip']

sizes = 50 * data['tip']
color = data['tip']
plt.figure(figsize=(10, 7))

plt.scatter(x, y, s = sizes,c= color, alpha = 0.5, cmap = "viridis")
plt.xlabel("total_bill")
plt.ylabel("tip")
plt.title("total_bill & tip Scatter plot")
plt.colorbar()
```

## 2. 성별에 따른 tip 비교하기 (bar plot)

```python
male_tip = data[data['sex'] == "Male"]['tip']
female_tip = data[data['sex']== "Female"]['tip']

r1 = male_tip.index
r2 = female_tip.index

plt.figure(figsize=(20,12))

plt.bar(r1,male_tip, width = 0.8, label = 'Male', color = 'forestgreen')
plt.bar(r2, female_tip, width = 0.8, label = 'Female', color = 'hotpink')

plt.xlabel("Group",size = 24)
plt.ylabel("tip",size=24)
plt.legend(fancybox = True, framealpha = 1, shadow =True, borderpad = 1)
```

## 3. 남자의 여자의 팁 금액의 분포도(histogram)를 그려 비교하세요

### 2 X 1 의 subplot을 이용하세요.

```python
male_tip = data[data['sex'] == "Male"]['tip']
female_tip = data[data['sex']== "Female"]['tip']

plt.figure(figsize = (20, 12))
plt.subplot(211)
plt.hist(male_tip, bins = 30, alpha = 0.6, color = "skyblue")
plt.xlim((0, 11))
plt.title("male tip", size = 24)

plt.subplot(212)
plt.hist(female_tip, bins = 30, alpha = 0.6 , color = "hotpink")
plt.title("female tip", size = 24)
plt.xlim(0, 11)
plt.ylim(0, 34);
```
